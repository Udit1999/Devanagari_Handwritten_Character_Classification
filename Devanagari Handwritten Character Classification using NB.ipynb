```python
X_Train_Data = Train_DataFrame.iloc[:,1:]
Y_Train_Data = Train_DataFrame.iloc[:,0]

X_Test_Data = Test_DataFrame.iloc[:,1:]
Y_Test_Data = Test_DataFrame.iloc[:,0]
```

```python
def PCA(Train_Data,Test_Data):
    Cov = np.cov(Train_Data.T)
    eigen_pairs = {}
    eigen_vals,eigen_vecs = np.linalg.eig(Cov)

    for i in range(len(eigen_vals)):
        eigen_pairs[eigen_vals[i]] = eigen_vecs[i] 

    tot = sum(eigen_vals)

    var_exp = [(i/tot) for i in sorted(eigen_vals,reverse= True)]
    cum_var_exp = np.cumsum(var_exp)


    for i in range(len(cum_var_exp)):
        if cum_var_exp[i] >= 0.999:
            k = i
            break;

    print("K :",k)
    print("LEN: ",len(eigen_vals))


    sorted(eigen_vals,reverse=True)


    w = []
    for i in eigen_vals:
        w.append(np.real(eigen_pairs[i][:k]))

    w = np.array(w)


    print("w.shape",w.shape)
    X_Train_Data_PCA = Train_Data.dot(w)
    print("X_Train_Data_PCA.shape",X_Train_Data_PCA.shape)

    X_Test_Data_PCA = Test_Data.dot(w)
    print("X_Test_Data_PCA.shape",X_Test_Data_PCA.shape)
    
    return X_Train_Data_PCA,X_Test_Data_PCA
```

```python
PCapList =[Y_Train_Data[Y_Train_Data == i].shape[0]/Y_Train_Data.shape[0] for i in labels]
PCapList
```

```python
CharacterList = {}
for label in labels:
    CharacterList[label] = X_Train_Data[Y_Train_Data == label]
    print(CharacterList[label].shape)
```

```python
from math import log
from scipy.stats import norm
```

```python
def calProb(Test_Data_PCA,Pcap,Mean,cov):
    return log(Pcap) + s.multivariate_normal.logpdf(Test_Data_PCA,Mean,cov)


ProbList = []
lenght_list = range(len(labels))

for label,i in zip(labels,lenght_list):
    X_Train_Data_PCA,X_Test_Data_PCA = PCA(CharacterList[label],X_Test_Data)
    Mean = np.mean(a = X_Train_Data_PCA,axis=0)
    print(Mean.shape)
    Cov = np.cov(X_Train_Data_PCA.T)
    print(Cov.shape)
    prob = calProb(X_Test_Data_PCA,PCapList[i],Mean,Cov)
    print("\n\n",prob)
    print(prob.shape)
    ProbList.append(prob)

```

```python
ProbList = np.array(ProbList)
problist = ProbList.T
print(problist.shape)
```
